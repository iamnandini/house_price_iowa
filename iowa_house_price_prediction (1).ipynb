Import Dependencies


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Path of the file to read
iowa_file_path = '/content/drive/MyDrive/Colab_Notebooks/train.csv'


#This reads CSV file
home_data = pd.read_csv(iowa_file_path)
```

Understanding Data

```python
home_data.describe()
```

Defining Features and Targets

```python
# Creating target object and calling it y
y = home_data.SalePrice
# Creating X
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]
```

```python
# printing description or statistics from X
print(X.describe())

# printing the top few lines
print(X.head())
```

```python
# Spliting into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

Training and Testing Models

```python
# Specifing Model
iowa_model = DecisionTreeRegressor(random_state=1)
# Fiting Model
model = iowa_model.fit(train_X, train_y)

# Make validation predictions and calculate mean absolute error
val_predictions = model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))

# Using best value for max_leaf_nodes
iowa_model = DecisionTreeRegressor(max_leaf_nodes=100, random_state=1)
iowa_model.fit(train_X, train_y)
val_predictions1 = iowa_model.predict(val_X)


val_mae1 = mean_absolute_error(val_predictions1, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae1))


# Calculating the Mean Absolute Error in Validation Data
val_mae = mean_absolute_error(val_predictions, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))

mape = mean_absolute_percentage_error(val_y, val_predictions) * 100
print ("{:,.0f}%".format(mape))

```

```python
rf_model = RandomForestRegressor(random_state=1)
# fit your model
model = rf_model.fit(train_X,train_y)

rf_prediction = model.predict(val_X)

# Calculate the mean absolute error of your Random Forest model on the validation data
rf_val_mae = mean_absolute_error(rf_prediction,val_y)
print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))

mape = mean_absolute_percentage_error(val_y, rf_prediction) * 100
print ("{:,.0f}%".format(mape))
```
